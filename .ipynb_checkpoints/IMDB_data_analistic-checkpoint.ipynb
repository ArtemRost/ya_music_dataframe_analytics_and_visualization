```python
import pandas as pd 
import seaborn as sns
import matplotlib as pl
import numpy as np

df = pd.read_csv('data.csv')
df = df.rename(columns={'vote_average':'user_score'})
```

```python
df[['director','budget','runtime']].sort_values(by='budget', ascending=False).head(10)
```

```python
# успех фильма очень сильно зависит от его бюджета. давайте проверим и узнаем это
# какие метрики зависят от бюджета
#
#
```

```python
mat_cor = df[['budget', 'revenue','popularity','runtime','vote_count','user_score','release_year']]
```

```python
matr = np.triu(mat_cor.corr())

sns.heatmap(mat_cor.corr(), annot = True, vmin = -1, vmax = 1, center = 0, mask=matr)
```

```python
# визуализируем матрицу корреляции с помощью heatmaps, чтобы понять зависимости разных данных друг от друга.
# из графикка можно сделать вывод что от бюджета фильма зависит его популярность, так как 
# бюджет фильма подразумевает затраты на его продвижение и т.д.
# также от бюджета зависит выручка фильма, но выручка ещё тесно связанна с популярностью, 
# так как чем больше людей узнают о фильме, тем больше людей придёт в кинотеатры
# длина фильма уже чуть меньше зависит от бюджета, но очевидно, что снять длинный фильм дороже
# но из матрицы корреляции можно сделать вывод о том, что количество оценок фильма сильно связано
# с его популярностью и его выручкой. 
#
#
#
```
