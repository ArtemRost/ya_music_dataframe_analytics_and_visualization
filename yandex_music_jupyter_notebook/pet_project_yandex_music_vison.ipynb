## Заметим, что реп стремительно набирал обороты во всем мире. Давайте проанализируем какие жанры встречаются в топ 100 треков яндекс музыки.

```python
top = df['genre'].value_counts().head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top.values, y=top.index, hue=top.values,palette='pastel',legend=False)
plt.xlabel('Количество треков')
plt.ylabel('Жанры')
plt.title('Топ 10 жанров в топе Яндекс музыки')
plt.tight_layout()
plt.show()
```

## Заметим, что Россия тоже не отстает и реп в ней стал ведущим жанром, хотя пятнадцать лет назад реп много кто даже за жанр не считал

```python
grouped = df.groupby('genre')
agg_functions = {
    'monthly_listens_total': 'sum',
    'artists_likes_total': 'sum'
}
genre_stats = grouped.agg(agg_functions)
genre_listens = genre_stats['monthly_listens_total'].sort_values(ascending=False)
genre_listens

plt.figure(figsize=(12, 6),dpi=600)
sns.barplot(x=genre_listens.values, y=genre_listens.index,hue=genre_listens.values ,palette='pastel',legend=False)
plt.xlabel('Прослушивания за один месяц')
plt.ylabel('Жанры')
plt.title('Топ 10 жанров по прослушиваниям за месяц')
plt.tight_layout()
plt.show()
```

```python
grouped = df['artist(s)'].value_counts().head(10)

plt.figure(figsize=(12, 6),dpi=600)
sns.barplot(x=grouped.values, y=grouped.index,hue=grouped.values ,palette='pastel',legend=False)
plt.xlabel('Количество треков в чарте')
plt.ylabel('Артисты')
plt.title('Топ количества треков артистов в чарте Яндекс Музыки')
plt.tight_layout()
plt.show()
```

## Из графиков можно заметить, что из-за нашумевшего сериала "Король и Шут" произошло возрождение творчества группы, что наблюдается в топе количества треков в чарте и в топе жанров в чарте

### Давайте построим график самых популярных треков с учетом запрещеного контента в песне и посмотрим как он влияет на то, сколько людей его послушают

```python
most_pop_track = df.sort_values(by='artists_likes_total', ascending=False)


plt.figure(figsize=(23, 15))
sns.scatterplot(data=most_pop_track, x='monthly_listens_total', y='artists_likes_total', hue='Explicit_content', palette="deep")

for i, row in most_pop_track.head(10).iterrows():  
    plt.annotate(row['name'], (row['monthly_listens_total'], row['artists_likes_total']),
                 fontsize=10, xytext=(-10, 10), textcoords='offset points', ha='center')

plt.xlabel('Прослушивания за месяц')
plt.ylabel('Лайки артиста')
plt.title('Самые популярные треки')
plt.tight_layout()

plt.show()
```

## Из этого графика можно сделать вывод, что содержание запрещенного контента в песне не влияет на его популярность и рейтинг артиста. Скорее можно подтвердить очевидную гипотезу, что треки, в которых нет запрещенного контента, охватывают большую группу людей и соответственно популярность выше
